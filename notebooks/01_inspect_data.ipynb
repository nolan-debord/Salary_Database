```python
# Remove the State level rows from the salary_df dataset
# Will only keep rows where the AREA_TITLE contains a , like "Houston, TX"
# Will eliminate "USA", "TX" and other single state/country references
salary_df = salary_df[salary_df['AREA_TITLE'].str.contains(",")].copy()

# Seperate the city, state into two seperate columns for better matching
salary_df[['City', 'State']] = salary_df['AREA_TITLE'].str.split(",", expand=True)

# Explode MSAs into separate rows for each city
salary_df = salary_df.copy()
salary_df['City'] = (
    salary_df['City']
    .str.split('-')  # Split "Denver-Aurora-Centennial" â†’ ['Denver', 'Aurora', 'Centennial']
)

# Explode creates multiple rows per MSA, one per city
salary_df = salary_df.explode('City')
salary_df['City'] = salary_df['City'].str.strip().str.title()

# Remove any city after the initial City in some of the longer - ones
salary_df['City'] = salary_df['City'].str.split('-').str[0].str.strip().str.title()
salary_df['State'] = salary_df['State'].str.split('-').str[0].str.strip().str.upper()

# Columns to use
# Select occupation name and annual salary percentiles
annual_cols = ['TOT_EMP', 'LOC_QUOTIENT', 'A_MEAN', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90']
salary_clean = salary_df[['City', 'State', 'OCC_TITLE'] + annual_cols].copy()

# Remove commas and convert to numeric for annual salary columns
for col in annual_cols:
    salary_clean[col] = (
        salary_clean[col]
        .astype(str)
        .str.replace(',', '', regex=False)  # Remove commas
    )
    salary_clean[col] = pd.to_numeric(salary_clean[col], errors='coerce')

# Verify changes
salary_clean[['City', 'State']].head(10)
```

### Impute data for cities where we don't have any salary/housing data

#### Helper function for distance calc

```python
def impute_missing_values_by_column(master_df, source_mask, column_name, k=1):
    """
    For a single column, impute missing values using nearest neighbors
    based on Latitude and Longitude.
    Also returns a mask marking which values were filled.
    """
    from sklearn.neighbors import NearestNeighbors

    df_with = master_df[source_mask & master_df[column_name].notna()]
    df_missing = master_df[master_df[column_name].isna()].copy()

    if df_with.empty or df_missing.empty:
        return master_df, pd.Series([False] * len(master_df), index=master_df.index)

    # Fit nearest neighbors
    nn = NearestNeighbors(n_neighbors=k)
    nn.fit(df_with[['Latitude', 'Longitude']])

    distances, indices = nn.kneighbors(df_missing[['Latitude', 'Longitude']])

    filled_mask = pd.Series([False] * len(master_df), index=master_df.index)

    for i, row_idx in enumerate(df_missing.index):
        nearest_index = df_with.index[indices[i][0]]
        master_df.at[row_idx, column_name] = master_df.at[nearest_index, column_name]
        filled_mask.at[row_idx] = True

    return master_df, filled_mask

```

#### Imputation

```python
salary_cols = [col for col in master_city_df.columns if col.startswith("Data Scientists_") or col.startswith("Registered Nurses_")]
housing_cols = ['6/30/2025', 'Cost of Living Index', 'Rent Index', 
                'Cost of Living Plus Rent Index', 'Groceries Index', 
                'Restaurant Price Index', 'Local Purchasing Power Index']

# Dictionary to store imputation flags
imputed_flags = {}

# Impute salary columns
for col in salary_cols:
    master_city_df, imputed_flags[col + "_imputed"] = impute_missing_values_by_column(
        master_city_df,
        source_mask=master_city_df['_salary_match'] == 'both',
        column_name=col
    )

# Impute housing columns
for col in housing_cols:
    master_city_df, imputed_flags[col + "_imputed"] = impute_missing_values_by_column(
        master_city_df,
        source_mask=master_city_df['_housing_match'] == 'both',
        column_name=col
    )

# Append flags to the master_city_df
for flag_col, flag_series in imputed_flags.items():
    master_city_df[flag_col] = flag_series
```

#### Merge the dataframes into one singular dataset to be used for mapping

```python
# Add Affordability metric

# Ensure the numeric columns are valid (if not already done)
cols_to_check = [
    'Data Scientists_A_MEAN', 'Registered Nurses_A_MEAN',
    '6/30/2025', 'Cost of Living Index'
]
for col in cols_to_check:
    master_city_df[col] = pd.to_numeric(master_city_df[col], errors='coerce')

# Compute total salary and affordability
master_city_df['Total_Salary'] = (
    master_city_df['Data Scientists_A_MEAN'].fillna(0) +
    master_city_df['Registered Nurses_A_MEAN'].fillna(0)
)

# Individual affordability
master_city_df['Affordability_DataScientist'] = (
    master_city_df['Data Scientists_A_MEAN'] / 
    (master_city_df['6/30/2025'] * master_city_df['Cost of Living Index'])
)

master_city_df['Affordability_RN'] = (
    master_city_df['Registered Nurses_A_MEAN'] / 
    (master_city_df['6/30/2025'] * master_city_df['Cost of Living Index'])
)

# Combined affordability
master_city_df['Affordability_Combined'] = (
    master_city_df['Total_Salary'] / 
    (master_city_df['6/30/2025'] * master_city_df['Cost of Living Index'])
)
```

```python
# Save the combined_df to csv
master_city_df.to_csv('../data/combined_affordability.csv', index=False)
```
